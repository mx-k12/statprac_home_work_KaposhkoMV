## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```
<img src="https://i.ibb.co/W4pyPfTX/tangent-line-3-0.png" width="600"/>

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

def beauty_derivative_plot(x_0):
    #тут определяем функцию, производную и касательную
    x = np.linspace(-2*np.pi, 2*np.pi, 1000) 
    y = 2*np.sin(x) - np.cos(2*x)   
    y_d = 2*np.cos(x) +2*np.sin(2*x) 
    k = 2*np.cos(x_0) +2*np.sin(2*x_0)
    f_x_0 = 2*np.sin(x_0) - np.cos(2*x_0)
    df_x_0 = round(2*math.cos(x_0) +2*math.sin(2*x_0), 2)
    tan_x = np.linspace(x_0-1, x_0+1, 100)
    tan_y = f_x_0 + k*(tan_x-x_0)


    #рисуем график, чтобы он был максимально похож на картинку
    plt.figure(figsize=(9, 6))
    plt.title("Function, Derivative, and Tangent Line")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.plot(x, y, label = 'f(x) = 2sinx - cos2x', color = 'blue')
    plt.plot(x, y_d, label = "f'(x) = 2cosx + 2sinx2x",  linestyle='dashed', color = 'red')
    plt.plot(tan_x, tan_y, label = 'Tangent at x = ' + str(x_0), color = 'green')
    plt.plot([x_0, x_0], [0, f_x_0], color='purple', linestyle='dashed')  # Перпендикуляр на ось Y
    plt.plot([0, x_0], [f_x_0, f_x_0], color='purple', linestyle='dashed')
    plt.scatter([x_0], [f_x_0], color='black', zorder=3)
    arrow = FancyArrowPatch(
        (x_0 - 0.1, f_x_0 - 1.77),  # Начало стрелки (текст с производной, ниже)
        (x_0, f_x_0),  # Конец стрелки (черная точка)
        connectionstyle="arc3,rad=.08",  # Сделаем стрелку более вертикальной
        arrowstyle='-|>',  # Стиль стрелки
        color='black',
        linewidth=1.1,
        mutation_scale=10,
        zorder=10
    )
    plt.gca().add_patch(arrow)
    plt.text(x_0 - 1.5, f_x_0 - 1.74, "f'(" + str(x_0)+ ')  = ' +str(df_x_0), fontsize=12, verticalalignment='top', horizontalalignment='left')
    plt.legend(loc = 'upper right')
    
    plt.show()
    

beauty_derivative_plot(-2.00)
````

## Problem 2

The data in [populations.txt](https://drive.google.com/file/d/1Bq7qVzbZIOd058VobGE6iYjUcMFkbZMt/view?usp=drive_link) describes the populations of hares, lynxes and carrots in northern Canada during 20 years. Do the following exercises **without for-loops**.

- Plot the population size of each species for each year on a single graph. Add legend and axis labels to the plot.

````python
import matplotlib.pyplot as plt
import numpy as np
#считывание данных
txt = np.loadtxt("populations.txt")
years = txt[:, 0]
hares = txt[:, 1]
lynxes = txt[:, 2]
carrots = txt[:, 3]


# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(years, hares, label='Hares')
plt.plot(years, lynxes, label='Lynxes')
plt.plot(years, carrots, label='Carrots')

# Добавление легенды и подписей осей
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Hares, Lynxes, and Carrots Over Time')
plt.grid(True)
plt.show()


````

- Find which year each species had the largest population. Mark these years in the plot of populations.

````python
max_hares_year = years[np.argmax(hares)]
max_lynxes_year = years[np.argmax(lynxes)]
max_carrots_year = years[np.argmax(carrots)]

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(years, hares, label='Hares')
plt.plot(years, lynxes, label='Lynxes')
plt.plot(years, carrots, label='Carrots')

# Отметим года с максимальной популяцией
plt.scatter(max_hares_year, max(hares), color='blue', zorder=5)
plt.scatter(max_lynxes_year, max(lynxes), color='red', zorder=5)
plt.scatter(max_carrots_year, max(carrots), color='green', zorder=5)

# Подписи и оформление
plt.legend()
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of Hares, Lynxes, and Carrots Over Time with Max Years')
plt.grid(True)
plt.show()
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each specis for each year in the period.

````python
plt.figure(figsize=(10, 6))
plt.bar(years, hares, label='Hares', color='blue')
plt.bar(years, lynxes, label='Lynxes', color='orange', bottom=hares)

# Оформление графика
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Stacked Bar Graph of Hares and Lynxes')
plt.legend()
plt.show()
````

- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
import numpy as np

# Средние значения
hares_mean = np.mean(hares)
lynxes_mean = np.mean(lynxes)
carrots_mean = np.mean(carrots)

# Построение гистограмм
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].hist(hares, bins=10, color='blue', alpha=0.7)
axes[0].axvline(hares_mean, color='black', linestyle='dashed', linewidth=2)
axes[0].set_title('Hares Population')
axes[0].set_xlabel('Population')
axes[0].set_ylabel('Frequency')

axes[1].hist(lynxes, bins=10, color='red', alpha=0.7)
axes[1].axvline(lynxes_mean, color='black', linestyle='dashed', linewidth=2)
axes[1].set_title('Lynxes Population')
axes[1].set_xlabel('Population')
axes[1].set_ylabel('Frequency')

axes[2].hist(carrots, bins=10, color='green', alpha=0.7)
axes[2].axvline(carrots_mean, color='black', linestyle='dashed', linewidth=2)
axes[2].set_title('Carrots Population')
axes[2].set_xlabel('Population')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
````

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

````python
# Сравнение популяций за каждый год
population_comparison = ['H' if hares[i] > lynxes[i] else 'L' for i in range(len(years))]

# Pie chart
plt.figure(figsize=(7, 7))
plt.pie([population_comparison.count('H'), population_comparison.count('L')],
        labels=['Hares', 'Lynxes'], autopct='%1.1f%%', startangle=90)
plt.title('Hares vs Lynxes Population Comparison')
plt.show()

print(population_comparison)
````

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
# Построение графика изменения популяции
plt.figure(figsize=(10, 6))
plt.plot(years, hares, label='Hares')
plt.plot(years, lynxes, label='Lynxes')

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Change Over Time for Hares and Lynxes')
plt.legend()
plt.grid(True)
plt.show()

# Корреляция
correlation = np.corrcoef(hares, lynxes)[0, 1]
print('Correlation coefficient between hare and lynx populations:'+ str(correlation))
````

- Assume the population of hares in 1920 is unknown. Suggest a way to estimate this value. Compare an estimated value
with the true value and print a ratio of the error to the true value.

````python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression



true_hares_1920 = hares[-1]  # Истинное значение, которое мы не используем в обучении

# Преобразуем списки в numpy-массивы
X_train = np.array(years[:-1]).reshape(-1, 1)  # Годы (признак)
y_train = np.array(hares[:-1])  # Численность зайцев (целевая переменная)

# Обучаем линейную регрессию
model = LinearRegression()
model.fit(X_train, y_train)

# Предсказываем популяцию зайцев в 1920 году
year_1920 = np.array([[1920]])
predicted_hares = model.predict(year_1920)[0]

# Вычисляем относительную ошибку
error_ratio = abs(predicted_hares - true_hares_1920) *100/ true_hares_1920 

# Вывод результатов
print('Предсказанная популяция зайцев в 1920 году: '+ str(int(predicted_hares)))
print('Истинная популяция зайцев в 1920 году: ' +str(true_hares_1920))
print('Относительная ошибка: '+str(round(error_ratio,2)))

````
